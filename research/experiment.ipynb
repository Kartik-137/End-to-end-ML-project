### Train, Test split

```python
# Separate features and target variable

x = df.drop("quality", axis=1)
y = df["quality"]
```

```python
# split the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Create the model
elastic_net = ElasticNet(alpha=0.5, l1_ratio=0.5, random_state=42)

# Fit the model
elastic_net.fit(x_train, y_train)
```

### Evaluation Metrics

```python
# Create a function to return all evaluation metrics

def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(y_true=actual, y_pred=pred))
    mae = mean_absolute_error(y_true=actual, y_pred=pred)
    r2 = r2_score(y_true=actual, y_pred=pred)

    return rmse, mae, r2
```
