Why use `@ensure annotations`

```python
def get_product(x:int, y:int) -> int:
    return x*y
```

```python
get_product(x = 2, y=3)

# Works because x and y are integer input 
```

```python
get_product(x = 2, y="3")

# Does work, but not as expected, should be 6
# Such issues are prevented using type checking libraries like `ensure_annotations`
```

```python
from ensure import ensure_annotations
@ensure_annotations
def get_product(x:int, y:int) -> int:
    return x*y

print(get_product(x = 2, y=3))
```

```python
# Provides an error because y is not integer
## A bug free function

print(get_product(x = 2, y="3"))
```
